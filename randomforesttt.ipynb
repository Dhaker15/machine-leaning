```
filterdata=df.iloc[0:,2:].values
```

```
filterdata
```

```
features=filterdata[0:,0:2]
```

```
label=filterdata[0:,2]
```

```
from sklearn.model_selection import train_test_split 
```

```
X,Y,x,y=train_test_split(features,label,test_size=0.2)
```

```
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
X=sc.fit_transform(X)
```

```
X
```

```
Y=sc.transform(Y)
```

```
Y
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
Rclf=RandomForestClassifier(n_estimators=20)
```

```
trained=Rclf.fit(X,x)
```

```
output=trained.predict(Y)
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y,output)
```
